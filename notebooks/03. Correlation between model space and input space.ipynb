```python
import datamol as dm
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from mood.dataset import MOOD_DATASETS
```

```python
PATH = "gs://experiments-output/mood-v2/results/dataframes/compare_spaces/20221216/correlations.csv"
assert dm.fs.exists(PATH)
```

```python
df = pd.read_csv(PATH)
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
sns.boxplot(data=df, x="dataset", y="spearman", order=MOOD_DATASETS, ax=ax)
ax.set_ylabel("Spearman Coefficient")
ax.set_xlabel("Dataset")
ax.set_xticklabels(ax.get_xticklabels(), rotation=15);
```
